## Setup

Run the cell below first. It loads the simulated vendor libraries. You do not need to read or modify this code.

```python
#@title 🔧 Setup — run this cell first (do not modify)
# This cell simulates the two vendor SDK libraries with realistic responses.
# In a real classroom, you would install the actual vendor libraries.
# For this study, the simulation lets you run the code without a physical drone.

import time

# === CoDrone EDU vendor library (simulated) ===
class _CodroneEDU_Drone:
    """Simulates the codrone_edu.drone.Drone class."""
    def __init__(self):
        self._battery = 87
        self._flying = False

    def pair(self):
        time.sleep(0.2)
        print("[CoDrone EDU] Paired with drone.")
        return True

    def takeoff(self):
        time.sleep(0.3)
        self._flying = True
        print("[CoDrone EDU] Taking off.")

    def hover(self, duration):
        time.sleep(0.2)
        print(f"[CoDrone EDU] Hovering for {duration} seconds.")

    def get_battery(self):
        return self._battery

    def land(self):
        time.sleep(0.3)
        self._flying = False
        print("[CoDrone EDU] Landing.")

    def close(self):
        print("[CoDrone EDU] Connection closed.")


# === Coding Rider vendor library (simulated) ===
# Note: this library has DIFFERENT method names and conventions
class _CodingRider_Drone:
    """Simulates the CodingRider.drone.Drone class."""
    def __init__(self):
        self._battery_level = 92
        self._is_flying = False

    def open(self):
        time.sleep(0.2)
        print("[Coding Rider] Connection opened.")
        return True

    def sendTakeOff(self):
        time.sleep(0.3)
        self._is_flying = True
        print("[Coding Rider] Sending takeoff command.")

    def eventState(self):
        # Returns a dict with battery info nested inside
        return {"battery": self._battery_level, "mode": "flight", "armed": True}

    def sendLanding(self):
        time.sleep(0.3)
        self._is_flying = False
        print("[Coding Rider] Sending landing command.")

    def close(self):
        print("[Coding Rider] Connection closed.")


# Make the mocks accessible as if they were the real imports
class _CodroneEDU_Module:
    Drone = _CodroneEDU_Drone

class _CodingRider_Module:
    Drone = _CodingRider_Drone

codrone_edu_drone = _CodroneEDU_Module()
CodingRider_drone = _CodingRider_Module()

print("✅ Setup complete. You may now proceed.")
```

## Vendor API Reference

The two vendor libraries expose **different method names** for the same operations.

### CoDrone EDU (`codrone_edu_drone.Drone`)
| Action | Method |
|---|---|
| Pair / connect | `.pair()` |
| Take off | `.takeoff()` |
| Hover in place | `.hover(duration)` |
| Read battery | `.get_battery()` → returns an integer (percentage) |
| Land | `.land()` |
| Close connection | `.close()` |

### Coding Rider (`CodingRider_drone.Drone`)
| Action | Method |
|---|---|
| Open connection | `.open()` |
| Take off | `.sendTakeOff()` |
| Hold position | -|
| Read state | `.eventState(state)` → returns a dict; battery is at `["battery"]` |
| Land | `.sendLanding()` |
| Close connection | `.close()` |
