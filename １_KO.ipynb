## 통합 API

프레임워크는 어떤 드론 제조사든 동일한 방식으로 작동하는 단일하고 일관된 API를 제공합니다.

| 개념 | 사용법 |
|---|---|
| 매니저 생성 | `manager = DroneManager()`|
| 플릿에 드론 추가 | `manager.add_drone(drone_id, adapter_key)` — 여기서 `adapter_key`는 `"mock_codrone_edu"` 또는 `"mock_coding_rider"`|
| 드론 한 대에 명령 전송 | `.hover(seconds)` |
| 배터리 읽기 | `manager.send_command(drone_id, DroneCommand.action(**params))` |
| 모든 드론에 동시에 명령 전송 | `manager.broadcast_command(DroneCommand.action(**params))` |

**사용 가능한 액션 (DroneCommand)**

- `DroneCommand.takeoff()`
- `DroneCommand.hover(duration) — takes `duration` parameter (in seconds)
- `DroneCommand.get_battery()`
- `DroneCommand.land()`
