## Framework API Reference

The framework exposes a single, consistent API that works the same way for any drone vendor.

| Concept | Usage |
|---|---|
| Create the manager | `manager = DroneManager()` |
| Add a drone to the fleet | `manager.add_drone(drone_id, adapter_key)` where `adapter_key` is `"mock_codrone_edu"` or `"mock_coding_rider"` |
| Send a command to one drone | `manager.send_command(drone_id, DroneCommand.action(**params))` |
| Send a command to ALL drones at once | `manager.broadcast_command(DroneCommand.action(**params))` |

### Available actions (`DroneCommand`)
- `DroneCommand.takeoff()`
- `DroneCommand.hover(duration) — takes `duration` parameter (in seconds)
- `DroneCommand.get_battery()`
- `DroneCommand.land()`
